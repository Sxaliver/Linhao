```python
import pyodbc
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = r"153 datasets/avall.mdb"
conn_str = r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=' + file_path + ';'

# 连接数据库
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

# 获取所有表名
table_names = [table.table_name for table in cursor.tables(tableType='TABLE')]
print("数据库中的表名：", table_names)

# 关闭连接
cursor.close()
conn.close()
```

```python
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

# 获取所有表的名称
table_names = [table.table_name for table in cursor.tables(tableType='TABLE')]

# 打印每个表的字段信息
for table in table_names:
    columns = [column.column_name for column in cursor.columns(table=table)]
    print(f"\n表 '{table}' 的字段信息：", columns)

# 关闭连接
cursor.close()
conn.close()
```

```python
# 连接数据库
conn = pyodbc.connect(conn_str)

# 查询 on_ground_collision 列中各唯一值的条数
query = """
SELECT on_ground_collision, COUNT(*) as count
FROM events
GROUP BY on_ground_collision
"""
df = pd.read_sql(query, conn)

print("on_ground_collision 列中各唯一值的条数：\n", df)

# 关闭连接
conn.close()
```

```python
# 连接数据库
conn = pyodbc.connect(conn_str)

# 查询 on_ground_collision 列中值为 Y 的行
query = "SELECT * FROM events WHERE on_ground_collision = 'Y' OR on_ground_collision IS NULL"
df = pd.read_sql(query, conn)

print("collision number  = ", len(df))  
print("events 表中 on_ground_collision 列值不为 N 的行：\n", df)  # 显示前几行数据
import pandas as pd

# save
df.to_excel('collisions.xlsx', index=False)

# 关闭连接
conn.close()
```
